## 数据类型 
编程中有数据类型（ data type）这一概念。数据类型表示数据的性质，有整数、小数、字符串等类型。Python中的``type()``函数可以用来查看数据类型。

````python
#数据类型
type(10)#<class 'int'>
type(2.718)#<class 'float'>
type("hello")#<class 'str'>
type([1,2,3])#<class 'list'>
````

## 变量 
可以使用x或y等字母定义变量（variable）。此外，可以使用变量进行计算，也可以对变量赋值。

````python
#变量
x=10
print(x)#10
y=3.14
print(x*y)#31.400000000000002
type(x*y)#<class 'float'>
````

## 列表 
除了单一的数值，还可以用列表（数组）汇总数据。

````python
#列表
a=[1,2,3,4,5]#列表
print(a)# [1, 2, 3, 4, 5]
len(a)#5
a[0]#1
a[4]#5
a[4]=99#修改列表元素
print(a)# [1, 2, 3, 4, 99]
````

元素的访问是通过``a[0]``这样的方式进行的。[]中的数字称为索引（下标），索引从0开始（索引0对应第一个元素）。此外，Python的列表提供了**切片**（slicing）这一便捷的标记法。使用切片不仅可以访问某个值，还可以访问列表的子列表（部分列表）。

````python
print(a)# [1, 2, 3, 4, 99]
a[0:2]#[1, 2]
a[1:]#获取第二个元素到最后一个元素[2, 3, 4, 99]
a[:3]#获取第一个元素到第三个元素[1, 2, 3]
a[:-1]#获取第一个元素到倒数第二个元素[1, 2, 3, 4]
a[:-2]#获取第一个元素到倒数第三个元素[1, 2, 3]
````

## 字典 
列表根据索引，按照0, 1, 2, ...的顺序存储值，而字典则以键值对的形式存储数据。字典就像《新华字典》那样，将单词和它的含义对应着存储起来。

````python
#字典，键值对，类似于java中的map
me={'height':180}#字典
me['height']#180
me['weight']=70
print(me)# {'height': 180, 'weight': 70}
````

## 布尔值 
Python中有bool型。bool型取``True``或``False``中的一个值。针对bool型的运算符包括and、or和not（针对数值的运算符有+、-、*、/等，根据不同的数据类型使用不同的运算符）。

````python
#布尔值
hungry=True#赋予变量hungry值为True
sleepy=False#赋予变量sleepy值为False
type(hungry)#<class 'bool'>
not hungry#False
hungry and sleepy#False
hungry or sleepy#True
````

## ``if``语句 
根据不同的条件选择不同的处理分支时可以使用``if/else``语句。 

Python中的空白字符具有重要的意义。上面的if语句中，``if hungry``:下面的语句开头有4个空白字符。它是缩进的意思，表示当前面的条件（``if hungry``）成立时，此处的代码会被执行。这个缩进也可以用tab表示，Python中推荐使用空白字符。

````python
# if语句
hungry=True
if hungry:
    print("I'm hungry")#缩进4个空白字符
else:
    print("I'm not hungry")
    print("I'm sleepy")
````

## ``for``语句 
进行循环处理时可以使用``for``语句。

这是输出列表[1, 2, 3]中的元素的例子。使用``for ... in ... ``:语句结构，可以按顺序访问列表等数据集合中的各个元素。

````python
#for语句
for i in [1,2,3]:
    print(i)

#range函数
for i in range(5):#0,1,2,3,4
    print(i)
````

## 函数 
可以将一连串的处理定义成函数（function）。

````python
#函数
#无参数的函数
def hello():
    print("Hello World!")
hello()#Hello World!

#带参数的函数
def hello(object):
    print("Hello "+object+"!")
hello("cat")#Hello cat!
````

## 类  
python中使用class关键字来定义类，类要遵循下述格式:  

```python
class 类名：  
    def __init__(self,参数,...):   #构造函数  
    ...  
    def 方法名1(self,参数,...):    #方法1  
    ...  
    def 方法名2(self,参数,...):    #方法2  
    ...  
```

这里有一个特殊的``__init__``方法，这是进行初始化的方法，也称为构造函数，只在生成类的实例时被调用一次。此外，在方法的第一个参数中明确地写入表示自身（自身的实例）的self是python的一个特点。


````python
#类，类是一种数据类型，类是一种抽象的数据类型，类是一种用户定义的数据类型(类似于C#中的类)
#类中的函数称为方法，类中的变量称为实例变量
class Man:
    def __init__(self,name,age):
        self.name=name#初始化实例变量，实例变量是在__init__方法中定义的，以self.变量名的方式定义
        self.age=age
        print("Initialized!")

    def hello(self):
        print("Hello "+self.name+"!")

    def goodbye(self):
        print("Good-bye "+self.name)

    def how_old(self):
        print("I'm "+str(self.age)+" years old")

#创建对象
m=Man("David",20)#调用__init__方法
m.hello()#调用hello方法 Hello David!
m.goodbye()#调用goodbye方法 Good-bye David
m.how_old()#调用how_old方法 I'm 20 years old
````

## ``Numpy``
在深度学习的实现中，经常出现数组和矩阵的计算。NumPy的数组类（``numpy.array``）中提供了很多便捷的方法，在实现深度学习时，我们将使用这些方法。本节我们来简单介绍一下后面会用到的NumPy。

要生成NumPy数组，需要使用``np.array()``方法。``np.array()``接收Python列表作为参数，生成NumPy数组（``numpy.ndarray``）

````python
#Numpy库,数值计算库,提供了数组和矩阵的运算,可以进行向量和矩阵的运算,可以进行线性代数运算
import numpy as np
x=np.array([1.0,2.0,3.0])
print(x)# [1. 2. 3.]
type(x)#<class 'numpy.ndarray'>
print(type(x))#<class 'numpy.ndarray'>
````

### ``NumPy``的算术运算

````python
#Numpy的算术运算
x=np.array([1.0,2.0,3.0])
y=np.array([2.0,4.0,6.0])
x+y#array([3., 6., 9.])
x-y#array([-1., -2., -3.])
x*y#array([ 2.,  8., 18.])
x/y#array([0.5, 0.5, 0.5])
x/2.0#array([0.5, 1. , 1.5])
x//2.0#array([0., 1., 1.])
````

### ``NumPy``的N维数组 
NumPy不仅可以生成一维数组（排成一列的数组），也可以生成多维数组。比如，可以生成如下的二维数组（矩阵）。

````python
#Numpy的N维数组
A=np.array([[1,2],[3,4]])
print(A)
A.shape#(2, 2)
A.dtype#dtype('int64')
B=np.array([[3,0],[0,6]])
print(A+B)#array([[ 4,  2],[ 3, 10]])
print(A*B)#算数点乘array([[ 3,  0],[ 0, 24]])
print(A*10)#array([[10, 20],[30, 40]])
#数学上将一维数组称为向量,二维数组称为矩阵，N维数组称为张量，这里将三维和三维以上的数组称为张量
````

## 广播 
NumPy中，形状不同的数组之间也可以进行运算。之前的例子中，在2×2的矩阵A和标量10之间进行了乘法运算。在这个过程中，如图1-1所示，标量10被扩展成了2 × 2的形状，然后再与矩阵A进行乘法运算。这个巧妙的功能称为广播（broadcast）。

````python
#广播（python的简化运算）
#广播是指不同形状的数组之间的运算，Numpy中的广播功能可以使形状不同的数组之间的运算变得简单
#广播的原则是，如果两个数组的后缘维度的轴长度相符或其中一方的长度为1，则认为他们是广播兼容的，广播会在缺失的轴上进行扩展，使得两个数组的维度相同，然后进行逐元素运算
A=np.array([[1,2],[3,4]])
B=np.array([10,20])
print(A*B)#array([[10, 40],[30, 80]])
````

## 访问元素 
元素的索引从0开始。对各个元素的访问可按如下方式进行。

````python
#访问元素
X=np.array([[51,55],[14,19],[0,4]])
print(X)
print()
print(X[0],end="\n\n")#第零行
print(X[0][1])#第零行第一列
print()
for row in X:#遍历行
    print(row)
print()
X=X.flatten()#将X转换为一维数组
print(X)
print()
print(X[np.array([0,2,4])])#获取索引为0,2,4的元素
print()
print(X>15)#返回一个布尔数组
print()
print(X[X>15])#返回一个一维数组，其中包含大于15的元素
````

## ``Matplotlib``
在深度学习的实验中，图形的绘制和数据的可视化非常重要。Matplotlib是用于绘制图形的库，使用Matplotlib可以轻松地绘制图形和实现数据的可视化。这里，我们来介绍一下图形的绘制方法和图像的显示方法。

````python
#Matplotlib库
import numpy as np
import matplotlib.pyplot as plt

#数据生成
x=np.arange(0,6,0.1)#生成0到6的数组，以0.1为单位
y=np.sin(x)

#绘制图形
plt.plot(x,y)
plt.show()
````

## ``pyplot``的功能
在刚才的``sin``函数的图形中，我们尝试追加``cos``函数的图形，并尝试使用pyplot的添加标题和x轴标签名等其他功能。

````python
#pyplot的功能
import numpy as np
import matplotlib.pyplot as plt

#数据生成
x=np.arange(0,6,0.1)
y1=np.sin(x)
y2=np.cos(x)

#绘制图形
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--",label="cos")#用虚线绘制
plt.xlabel("x")#x轴标签
plt.ylabel("y")#y轴标签
plt.title('sin & cos')#标题
plt.legend()#显示图例
plt.show()
````

## 显示图像

````python
#显示图像
#pyplot还提供了显示图像的方法imshow()。另外，可以使用matplotlib.image模块的imread()方法读入图像文件
import matplotlib.pyplot as plt
from matplotlib.image import imread#导入imread方法
img=imread('lena.png')#读入图像
plt.imshow(img)#显示图像
````
