## 2. 快速分析 (ReAct)

获取实时行情，约 30 秒完成。

```python
result = agent.react("贵州茅台600519现在什么价格？")
print(result)
```

## 3. 深度分析 (PlanSolve)

从基本面、技术面、估值等多维度全面评估，约 2-3 分钟。

```python
result = agent.plan_solve("全面分析中国平安601318的投资价值")
print(result)
```

## 4. 批判分析 (Reflection)

生成报告后再自我审查改进，约 3-5 分钟。

```python
result = agent.reflect("宁德时代300750目前是否值得买入？")
print(result)
```
